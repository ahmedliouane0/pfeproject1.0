```python
import pandas as pd
from itertools import product
import re
import unicodedata

languages = ['en_US','fr_FR']

def create_translation_pairs(df):

    pairs = []
    df_shuffled = df.dropna().sample(frac=1, random_state=42)

    for _, row in df_shuffled.iterrows():
        for src_lang, tgt_lang in product(languages, repeat=2):
            if src_lang != tgt_lang:
                source_text = f"<{src_lang}> " + str(row[src_lang]).strip()
                target_text = f"<{tgt_lang}> " + str(row[tgt_lang]).strip()

                pairs.append({
                    "source": source_text,
                    "target": target_text,
                    "source_lang": src_lang,
                    "target_lang": tgt_lang
                })

    return pd.DataFrame(pairs)

def preprocess_text(text, lowercase=True):
    """
    Normalize and clean text for better tokenization.
    Preserves the language tag at the beginning.
    """
    lang_tag_match = re.match(r'^(<[^>]+>)', text)
    lang_tag = lang_tag_match.group(1) if lang_tag_match else ""

    if lang_tag:
        text = text[len(lang_tag):].strip()

    text = re.sub(r'\s+', ' ', text).strip()

    text = unicodedata.normalize('NFKC', text)

    text = re.sub(r'([.,!?():])', r' \1 ', text)
    text = re.sub(r'\s+', ' ', text).strip()

    if lowercase:
        text = text.lower()

    if lang_tag:
        text = f"{lang_tag} {text}"

    return text

def generate_and_save_pairs(df_train, df_test, df_validation):
    """
    Generate, preprocess, and save multilingual translation pairs.
    """
    train_pairs = create_translation_pairs(df_train)
    test_pairs = create_translation_pairs(df_test)
    valid_pairs = create_translation_pairs(df_validation)

    for df in [train_pairs, test_pairs, valid_pairs]:
        df['source'] = df['source'].apply(preprocess_text)
        df['target'] = df['target'].apply(preprocess_text)

    print(f"Training pairs: {len(train_pairs)}")
    print(f"Testing pairs: {len(test_pairs)}")
    print(f"Validation pairs: {len(valid_pairs)}")

    train_pairs.to_csv("train_pairs.csv", index=False)
    test_pairs.to_csv("test_pairs.csv", index=False)
    valid_pairs.to_csv("valid_pairs.csv", index=False)

    return train_pairs, test_pairs, valid_pairs
```

```python
def tokenize_data(df, tokenizer):
    input_ids = []
    target_ids = []

    bos_id = tokenizer.bos_id()
    eos_id = tokenizer.eos_id()

    for _, row in df.iterrows():
        source_text = row['source']
        target_text = row['target']

        input_encoded = tokenizer.encode(source_text, out_type=int)
        target_encoded = tokenizer.encode(target_text, out_type=int)


        input_ids.append([bos_id] + input_encoded + [eos_id])
        target_ids.append([bos_id] + target_encoded + [eos_id])

    df['input_ids'] = input_ids
    df['target_ids'] = target_ids

    return df



train_pairs_tokenized = tokenize_data(final_train_pairs, sp)
test_pairs_tokenized = tokenize_data(final_test_pairs, sp)
valid_pairs_tokenized = tokenize_data(final_valid_pairs, sp)


train_pairs_tokenized.to_csv("train_tokenized.csv", index=False)
test_pairs_tokenized.to_csv("test_tokenized.csv", index=False)
valid_pairs_tokenized.to_csv("valid_tokenized.csv", index=False)
```

```python
import torch
from torch.utils.data import Dataset
from torch.nn.utils.rnn import pad_sequence

class TranslationDataset(Dataset):
    def __init__(self, df):
        self.input_ids = df['input_ids'].tolist()
        self.target_ids = df['target_ids'].tolist()

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        input_ids = torch.tensor(self.input_ids[idx])
        target_ids = torch.tensor(self.target_ids[idx])

        return {
            'input_ids': input_ids,
            'target_ids': target_ids
        }

def collate_fn(batch):
    pad_id = 0

    input_ids  = [item['input_ids']  for item in batch]
    target_ids = [item['target_ids'] for item in batch]

    input_lengths  = torch.tensor([len(x) for x in input_ids])
    target_lengths = torch.tensor([len(x) for x in target_ids])

    input_ids  = pad_sequence(input_ids,  batch_first=True, padding_value=pad_id)
    target_ids = pad_sequence(target_ids, batch_first=True, padding_value=pad_id)

    attention_mask = (input_ids != pad_id).long()

    return {
        'input_ids':      input_ids,
        'attention_mask': attention_mask,
        'target_ids':     target_ids,
        'input_lengths':  input_lengths,
        'target_lengths': target_lengths
    }


train_dataset = TranslationDataset(train_pairs_tokenized[:50000])
test_dataset = TranslationDataset(test_pairs_tokenized[:10000])
valid_dataset = TranslationDataset(valid_pairs_tokenized[:10000])

```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import random

class Encoder(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, n_layers=1, dropout=0.5):
        super().__init__()
        self.hidden_dim = hidden_dim
        self.n_layers = n_layers

        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.rnn = nn.LSTM(embed_dim, hidden_dim, n_layers,
                          bidirectional=True, batch_first=True, dropout=dropout if n_layers > 1 else 0)
        self.fc_hidden = nn.Linear(hidden_dim * 2, hidden_dim)
        self.fc_cell = nn.Linear(hidden_dim * 2, hidden_dim)
        self.dropout = nn.Dropout(dropout)

    def forward(self, src, src_lengths):
        embedded = self.dropout(self.embedding(src))

        packed_embedded = nn.utils.rnn.pack_padded_sequence(
            embedded, src_lengths.cpu(), batch_first=True, enforce_sorted=False)

        packed_outputs, (hidden, cell) = self.rnn(packed_embedded)
        outputs, _ = nn.utils.rnn.pad_packed_sequence(packed_outputs, batch_first=True)


        hidden = torch.cat((hidden[-2], hidden[-1]), dim=1)
        cell = torch.cat((cell[-2], cell[-1]), dim=1)

        hidden = torch.tanh(self.fc_hidden(hidden))
        cell = torch.tanh(self.fc_cell(cell))

        return outputs, (hidden, cell)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

class Attention(nn.Module):
    def __init__(self, hidden_dim, method='general'):
        super().__init__()
        self.method = method

        if method == 'general':
            self.attn = nn.Linear(hidden_dim * 2, hidden_dim)
        elif method == 'concat':
            self.attn = nn.Linear(hidden_dim * 3, hidden_dim)
            self.v = nn.Linear(hidden_dim, 1, bias=False)

    def forward(self, hidden, encoder_outputs, mask=None):
        if encoder_outputs.dim() == 2:
            encoder_outputs = encoder_outputs.unsqueeze(0)

        batch_size = encoder_outputs.size(0)
        src_len = encoder_outputs.size(1)

        if hidden.dim() == 3:
            hidden = hidden[-1]

        hidden = hidden.unsqueeze(1).repeat(1, src_len, 1)

        if self.method == 'general':
            energy = torch.tanh(self.attn(encoder_outputs))
            attention = torch.sum(hidden * energy, dim=2)
        elif self.method == 'concat':
            energy = torch.tanh(self.attn(torch.cat((hidden, encoder_outputs), dim=2)))
            attention = self.v(energy).squeeze(2)

        if mask is not None:
            attention = attention.masked_fill(mask == 0, -1e10)

        attn_weights = F.softmax(attention, dim=1)

        context = torch.bmm(attn_weights.unsqueeze(1), encoder_outputs)
        context = context.squeeze(1)

        return context, attn_weights
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import random


class Decoder(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, attention, n_layers=1, dropout=0.5):
        super().__init__()
        self.vocab_size = vocab_size
        self.hidden_dim = hidden_dim
        self.n_layers = n_layers
        self.attention = attention

        self.embedding = nn.Embedding(vocab_size, embed_dim)
        self.rnn = nn.LSTM(embed_dim + hidden_dim * 2, hidden_dim, n_layers,
                          batch_first=True, dropout=dropout if n_layers > 1 else 0)
        self.fc_out = nn.Linear(hidden_dim + hidden_dim * 2 + embed_dim, vocab_size)
        self.dropout = nn.Dropout(dropout)

    def forward(self, input, hidden, encoder_outputs, mask=None):
        if input.dim() == 1:
            input = input.unsqueeze(1)
        elif input.size(1) != 1:
            input = input.unsqueeze(1)

        embedded = self.dropout(self.embedding(input))

        hidden_state, cell_state = hidden

        context, attn = self.attention(hidden_state, encoder_outputs, mask)
        context = context.unsqueeze(1)

        rnn_input = torch.cat((embedded, context), dim=2)

        if hidden_state.dim() == 2:
            hidden_reshaped = hidden_state.unsqueeze(0).repeat(self.n_layers, 1, 1)
            cell_reshaped = cell_state.unsqueeze(0).repeat(self.n_layers, 1, 1)
            hidden = (hidden_reshaped, cell_reshaped)

        output, (new_hidden, new_cell) = self.rnn(rnn_input, hidden)
        output = output.squeeze(1)
        embedded = embedded.squeeze(1)
        context = context.squeeze(1)

        prediction = torch.cat((output, context, embedded), dim=1)
        prediction = self.fc_out(prediction)

        return prediction, (new_hidden[-1], new_cell[-1]), attn
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import random

class Seq2SeqModel(nn.Module):
    def __init__(self, encoder, decoder, device, pad_idx=0):
        super().__init__()
        self.encoder = encoder
        self.decoder = decoder
        self.device = device
        self.pad_idx = pad_idx

    def forward(self, src, src_lengths, trg, teacher_forcing_ratio=0.5):
        batch_size = src.shape[0]
        trg_len = trg.shape[1]
        trg_vocab_size = self.decoder.vocab_size

        outputs = torch.zeros(batch_size, trg_len-1, trg_vocab_size).to(self.device)

        encoder_outputs, (hidden, cell) = self.encoder(src, src_lengths)

        mask = (src != self.pad_idx).to(self.device)

        input = trg[:, 0]

        for t in range(1, trg_len):
            output, (hidden, cell), _ = self.decoder(input, (hidden, cell), encoder_outputs, mask)

            outputs[:, t-1] = output

            teacher_force = random.random() < teacher_forcing_ratio
            top1 = output.argmax(1)
            input = trg[:, t] if teacher_force else top1

        return outputs

    def create_mask(self, src):
        return (src != self.pad_idx).to(self.device)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

VOCAB_SIZE = 8000
EMBED_SIZE = 512
HIDDEN_SIZE = 640
NUM_LAYERS = 2
DROPOUT = 0.5
DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")

encoder = Encoder(VOCAB_SIZE, EMBED_SIZE, HIDDEN_SIZE, NUM_LAYERS, DROPOUT)
attn    = Attention(hidden_dim=HIDDEN_SIZE, method='general')

decoder = Decoder(
    vocab_size=VOCAB_SIZE,
    embed_dim=EMBED_SIZE,
    hidden_dim=HIDDEN_SIZE,
    attention=attn,
    n_layers=NUM_LAYERS,
    dropout=DROPOUT
)

model = Seq2SeqModel(encoder, decoder, DEVICE).to(DEVICE)
def init_weights(m):
    for name, param in m.named_parameters():
        if 'weight' in name:
            nn.init.xavier_uniform_(param.data)
        elif 'bias' in name:
            nn.init.constant_(param.data, 0)

model.apply(init_weights)

model.apply(init_weights)

PAD_IDX   = 0
criterion = nn.CrossEntropyLoss(ignore_index=PAD_IDX)
optimizer = optim.Adam(model.parameters(), lr=1e-4)


```

```python
def translate_batch(sentences, src_lang, tgt_lang, model, sp, device, beam_width=5, max_len=70, alpha=0.6):
    model.eval()
    translations = []

    bos_id = sp.bos_id()
    eos_id = sp.eos_id()
    pad_id = sp.pad_id()

    with torch.no_grad():
        for sentence in sentences:
            input_text = f"<{src_lang}> {sentence.strip()}"
            input_ids = sp.encode(input_text, out_type=int)
            input_tensor = torch.tensor(input_ids, dtype=torch.long).unsqueeze(0).to(device)
            input_length = torch.tensor([len(input_ids)], dtype=torch.long).to(device)

            encoder_outputs, hidden = model.encoder(input_tensor, input_length)
            mask = (input_tensor != pad_id).to(device)

            beams = [([bos_id], hidden, 0.0)]
            completed_beams = []

            for _ in range(max_len):
                new_beams = []

                for tokens, hidden_state, score in beams:
                    last_token = tokens[-1]

                    if last_token == eos_id:
                        completed_beams.append((tokens, score))
                        continue

                    decoder_input = torch.tensor([last_token], dtype=torch.long).to(device)

                    output, new_hidden, _ = model.decoder(decoder_input, hidden_state, encoder_outputs, mask)

                    log_probs = F.log_softmax(output, dim=1)
                    topk_log_probs, topk_indices = torch.topk(log_probs, beam_width)

                    for log_prob, idx in zip(topk_log_probs.tolist()[0], topk_indices.tolist()[0]):
                        new_tokens = tokens + [idx]
                        new_score = score + log_prob
                        new_beams.append((new_tokens, new_hidden, new_score))

                beams = sorted(new_beams, key=lambda x: x[2], reverse=True)[:beam_width]

                if len(completed_beams) >= beam_width:
                    break
            if completed_beams:
                best_tokens, best_score = max(completed_beams, key=lambda x: x[1] / (len(x[0]) ** alpha))
            else:
                best_tokens, _, _ = beams[0]

            clean_tokens = [tok for tok in best_tokens if tok not in [bos_id, eos_id, pad_id]]
            translated_text = sp.decode(clean_tokens)
            translated_text = translated_text.replace(f"<{tgt_lang}>", "").strip()
            translations.append(translated_text)

    return translations
```

```python
def translate_sentence(sentence, src_lang, tgt_lang, model, sp, device, max_len=100):
    model.eval()
    if src_lang:
        sentence = f"<{src_lang}> {sentence}"

    # tokenize + BOS/EOS
    tokens = sp.encode(sentence, out_type=int)
    tokens = [1] + tokens + [2]     # BOS=1, EOS=2
    src_tensor = torch.tensor(tokens, device=device).unsqueeze(0)
    src_length = torch.tensor([len(tokens)], device=device)

    with torch.no_grad():
        encoder_outputs, (hidden, cell) = model.encoder(src_tensor, src_length)
        mask = (src_tensor != 0).long()

        # **Don’t** expand here
        decoder_hidden, decoder_cell = hidden, cell

        input_tok = torch.tensor([1], device=device)  # BOS
        generated = []

        for _ in range(max_len):
            output, (decoder_hidden, decoder_cell), _ = model.decoder(
                input_tok, (decoder_hidden, decoder_cell), encoder_outputs, mask
            )
            top1 = output.argmax(1).item()
            if top1 == 2:  # EOS
                break
            generated.append(top1)
            input_tok = torch.tensor([top1], device=device)

    # Decode back to text
    output_text = sp.decode(generated)
    output_text = output_text.replace(f"<{src_lang}>", "").replace(f"<{tgt_lang}>", "").strip()

    # Return both translation and (empty) attention
    return output_text, None
```

```python
import re
import torch

# 1. Utility: Normalize Apostrophes
def normalize_apostrophes(text: str) -> str:
    return re.sub(r"\s+", " ",
        text.replace("’", " ' ").replace("'", " ' ")
    ).strip()

# 2. Beam Search Decoder with Length Penalty
def beam_search_decode(model, sp, enc_outs, enc_hidden, mask,
                       beam_width=5, max_len=100, alpha=0.6, device='cpu'):
    sequences = [[[], 0.0, enc_hidden]]  # (tokens_so_far, total_log_score, (h, c))

    def length_penalty(length, alpha=0.6):
        return ((5 + length) / 6) ** alpha

    for _ in range(max_len):
        all_candidates = []
        for seq, score, (h, c) in sequences:
            if seq and seq[-1] == sp.eos_id():
                all_candidates.append((seq, score, (h, c)))
                continue

            inp = torch.tensor([seq[-1]] if seq else [sp.bos_id()], device=device)
            out_logits, (h_new, c_new), _ = model.decoder(inp, (h, c), enc_outs, mask)
            log_probs = torch.nn.functional.log_softmax(out_logits, dim=1)
            topk = torch.topk(log_probs, beam_width)

            for i in range(beam_width):
                token_id = topk.indices[0][i].item()
                token_score = topk.values[0][i].item()
                candidate = (seq + [token_id], score + token_score, (h_new, c_new))
                all_candidates.append(candidate)

        # Apply length penalty before selecting top-k
        sequences = sorted(
            all_candidates,
            key=lambda tup: tup[1] / length_penalty(len(tup[0]), alpha),
            reverse=True
        )[:beam_width]

    return sequences[0][0]  # Return the best sequence (token list)

# 3. Translate a Single Sentence
def translate_text(text: str,
                   src_lang: str, tgt_lang: str,
                   model, sp, device,
                   max_len: int = 100, beam_width: int = 5, alpha: float = 0.6):
    model.eval()

    # Normalize and add source language tag
    txt = normalize_apostrophes(text)
    if src_lang:
        txt = f"<{src_lang}> {txt}"

    # Tokenize
    ids = sp.encode(txt, out_type=int)
    src = torch.tensor(ids, device=device).unsqueeze(0)
    lengths = torch.tensor([len(ids)], device=device)

    with torch.no_grad():
        enc_outs, (h, c) = model.encoder(src, lengths)
        mask = (src != sp.pad_id()).long()

        # Beam search decoding with length penalty
        gen_ids = beam_search_decode(model, sp, enc_outs, (h, c), mask,
                                     beam_width=beam_width, max_len=max_len, alpha=alpha, device=device)

    # Decode token IDs into text
    output_text = sp.decode_ids(gen_ids).strip()

    # Remove <lang> tag if it appears in the output
    if output_text.lower().startswith(f"<{tgt_lang.lower()}>"):
        output_text = output_text[len(f"<{tgt_lang}>"):].strip()

    return output_text.capitalize()

# 4. Translate a Batch of Sentences
def translate_batch(texts, src_lang, tgt_lang, **kwargs):
    return [translate_text(t, src_lang, tgt_lang, **kwargs) for t in texts]

# 5. Example Usage
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Add your trained model and tokenizer before running the below block
# model = ...  # your Seq2Seq model
# sp = ...     # your SentencePiece tokenizer

new_sentences = [
    "Where is the train station?",
    "I had already eaten when you arrived.",
    "It's raining cats and dogs.",
    "I don't like this movie.",
    "Can you help me?"
]

translations = translate_batch(
    new_sentences,
    src_lang="en_US",
    tgt_lang="fr_FR",
    model=model,
    sp=sp,
    device=device,
    beam_width=8,     # You can tweak beam width
    max_len=70,       # Optional limit on sentence length
    alpha=0.6         # Length penalty factor
)

for src, tgt in zip(new_sentences, translations):
    print(f"> {src}\n→ {tgt}\n")
```
